An alternative image normalization to histogram equalization is a quotient image.A quotient image is obtained by dividing the image with a blurred version I /(I ∗ Gσ ). Implement this and try it on some sample images.

```python
%matplotlib inline

from PIL import Image
from numpy import *
from pylab import *
from scipy.ndimage import filters
```

```python
im = array(Image.open('../images/empire.jpg').convert('L'))
print im.shape, im.dtype
```

```python
gray()
axis('equal')
axis('off')
imshow(im)

figure()
hist(im.flatten(), 128)
```

```python
blurred = filters.gaussian_filter(im, 5)
imshow(blurred)
figure()
hist(blurred.flatten(), 128)
```

```python
quotient_image = im / blurred * 255
imshow(quotient_image)
```

```python
hist(quotient_image.flatten(), 128)
```
