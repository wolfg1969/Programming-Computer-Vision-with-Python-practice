Take an image and apply Gaussian blur like in Figure 1-9. Plot the image contours for increasing values of σ . What happens? Can you explain why?

```python
%matplotlib inline

from PIL import Image
from pylab import *
from numpy import *
from scipy.ndimage import filters

def make_contour(im):
    figure()
    gray()
    contour(im, origin='image')
    axis('equal')
    axis('off')


def gaussian_filter(im, sigma):
    return filters.gaussian_filter(im, sigma)
    
gray()
```

### 原始图像的轮廓图

```python
im = array(Image.open('../images/empire.jpg').convert('L'))
make_contour(im)
title('original')
```

```python
### 高斯滤波 偏差值为 2 的轮廓图
```

```python
im2 = make_contour(gaussian_filter(im, 2))
title('gaussian filter with sigma=2')
```

```python
### 高斯滤波 偏差值为 5 的轮廓图
```

```python
im5 = make_contour(gaussian_filter(im, 5))
title('gaussian filter with sigma=5')
```

### 高斯滤波 偏差值为 10 的轮廓图

```python
im10 = make_contour(gaussian_filter(im, 10))
title('gaussian filter with sigma=10')
```

图像越模糊，细节越少，轮廓越明显。
